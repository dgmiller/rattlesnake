```python
Mentions = []
Hashtags = []
Weblinks = []
Retweets = []
for s in trump_words[:15]:
    mentions = re.findall(r'@\w*',s)
    hashtags = re.findall(r'#\w*',s)
    weblinks = re.findall(r'https\S*',s)#://[A-Za-z]*.\w*/\w*',s)
    retweets = re.findall('"RT ',s)
    Mentions.append(mentions)
    Hashtags.append(hashtags)
    Weblinks.append(weblinks)
    Retweets.append(retweets)
    for m in mentions:
        s = s.replace(m,"")
    for h in hashtags:
        s = s.replace(h,"")
    for w in weblinks:
        s = s.replace(w,"")
    for r in retweets:
        s = s.replace(r,"")
    print s.lower()
```
